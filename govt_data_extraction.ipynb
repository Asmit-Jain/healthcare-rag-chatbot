```python
def get_scheme_details(slug):

    url = (
        "https://api.myscheme.gov.in/"
        f"schemes/v6/public/schemes"
        f"?slug={slug}&lang=en"
    )

    response = requests.get(
        url,
        headers=HEADERS
    )

    return response.json()
```

```python
import json
import re
import time
import html
from tqdm import tqdm
from datetime import datetime

def clean_md_text(text):
    """
    Cleans raw Markdown by unescaping HTML entities, 
    converting <br> to newlines, and removing excess whitespace.
    """
    if not text:
        return ""
    
    # Unescape HTML entities (e.g., converts &amp;quot; to ")
    text = html.unescape(text)
    
    # Replace <br> tags with actual newlines
    text = re.sub(r'<br\s*/?>', '\n', text)
    
    # Strip leading/trailing whitespaces from each line
    lines = [line.strip() for line in text.split('\n')]
    text = '\n'.join(lines)
    
    # Remove excessive empty lines
    text = re.sub(r'\n{3,}', '\n\n', text)
    
    return text.strip()

def extract_text_from_rich_json(content_list):
    """
    Recursively digs into deeply nested JSON to extract text,
    automatically adding bullet points and newlines for readability.
    """
    extracted = ""
    if not isinstance(content_list, list):
        return extracted
        
    for item in content_list:
        if isinstance(item, dict):
            # Add a bullet point if the API says this is a list item
            prefix = "- " if item.get("type") == "list_item" else ""
            
            # Grab the direct text if it exists
            text_val = item.get("text", "")
            
            # If there are children, call this same function AGAIN (Recursion)
            children_text = ""
            if "children" in item:
                children_text = extract_text_from_rich_json(item["children"])
                
            combined = prefix + text_val + children_text
            
            # Add newlines after paragraphs, headers, and list items
            if item.get("type") in ["paragraph", "h1", "h2", "h3", "h4", "h5", "list_item"]:
                combined += "\n"
                
            extracted += combined
            
    return extracted
```

```python
def extract_single_scheme(scheme):
    """
    Fetches details, FAQs, and Documents for a single scheme and formats 
    them into the unified Document Store schema.
    """
    slug = scheme.get("slug")
    if not slug:
        return None
        
    # 1. Fetch Details API
    details_response = get_scheme_details(slug)
    if not details_response or details_response.get("status") != "Success":
        return None
        
    data_node = details_response.get("data") or {}
    scheme_id = data_node.get("_id")
    en_data = data_node.get("en") or {}
    
    # 2. Fetch FAQs API
    faqs_data = []
    if scheme_id:
        try:
            faqs_response = get_scheme_faqs(scheme_id)
            if faqs_response and faqs_response.get("status") == "Success":
                faqs_data = faqs_response.get("data", {}).get("en", {}).get("faqs", [])
        except Exception as e:
            print(f"Error fetching FAQs for {slug}: {e}")

    # 3. Fetch Documents API (UPDATED with Recursive Parser)
    docs_text = ""
    if scheme_id:
        try:
            docs_response = get_scheme_documents(scheme_id)
            if docs_response and docs_response.get("status") == "Success":
                docs_en = docs_response.get("data", {}).get("en", {})
                
                # Check for markdown first
                if docs_en.get("documents_required_md"):
                    docs_text = docs_en.get("documents_required_md")
                else:
                    # Unleash the recursive extractor on the deep JSON
                    docs_list = docs_en.get("documents_required", [])
                    docs_text = extract_text_from_rich_json(docs_list)
                    
        except Exception as e:
            print(f"Error fetching Documents for {slug}: {e}")
            
    # 4. Extract Content Sections
    content_sections = []
    
    # Overview
    if scheme.get("description"):
        content_sections.append({
            "section_title": "Overview",
            "text": clean_md_text(scheme.get("description"))
        })
        
    scheme_content = en_data.get("schemeContent", {})
    
    # Detailed Description
    if scheme_content.get("detailedDescription_md"):
        content_sections.append({
            "section_title": "Detailed Description",
            "text": clean_md_text(scheme_content.get("detailedDescription_md"))
        })
        
    # Benefits
    if scheme_content.get("benefits_md"):
        content_sections.append({
            "section_title": "Benefits",
            "text": clean_md_text(scheme_content.get("benefits_md"))
        })
        
    # Eligibility Criteria
    eligibility_node = en_data.get("eligibilityCriteria", {})
    if eligibility_node.get("eligibilityDescription_md"):
        content_sections.append({
            "section_title": "Eligibility Criteria",
            "text": clean_md_text(eligibility_node.get("eligibilityDescription_md"))
        })
        
    # Exclusions
    if scheme_content.get("exclusions_md"):
        cleaned_exc = clean_md_text(scheme_content.get("exclusions_md"))
        if cleaned_exc:
            content_sections.append({
                "section_title": "Exclusions",
                "text": cleaned_exc
            })

    # Application Process
    app_processes = en_data.get("applicationProcess", [])
    for idx, process in enumerate(app_processes):
        if process.get("process_md"):
            mode = process.get("mode", f"Method {idx+1}")
            content_sections.append({
                "section_title": f"Application Process ({mode})",
                "text": clean_md_text(process.get("process_md"))
            })

    # Documents Required 
    # Clean up double newlines that the recursive function might leave behind
    if docs_text.strip():
        clean_docs = re.sub(r'\n{3,}', '\n\n', docs_text)
        content_sections.append({
            "section_title": "Documents Required",
            "text": clean_md_text(clean_docs.strip())
        })
            
    # FAQs
    for faq in faqs_data:
        question = faq.get("question", "").strip()
        answer = faq.get("answer_md", "")
        if question and answer:
            content_sections.append({
                "section_title": f"FAQ: {question}",
                "text": clean_md_text(answer)
            })

    # 5. Return Structured Document
    return {
        "doc_id": f"gov_scheme_{slug.replace('-', '_')}",
        "title": scheme.get("title", "Unknown Scheme"),
        "source_name": "MyScheme (Government of India)",
        "source_url": f"https://www.myscheme.gov.in/schemes/{slug}",
        "category": "Government Healthcare Scheme",
        "topic": scheme.get("title", "Unknown Scheme"),
        "retrieved_on": datetime.now().strftime("%Y-%m-%d"),
        "content": content_sections
    }
```

```python
structured_schemes = []
seen_slugs = set() # Track unique schemes to avoid duplicates

for scheme in tqdm(all_schemes, desc="Processing Govt Schemes"):
    slug = scheme.get("slug")
    
    # Check for duplicates before hitting the API
    if not slug or slug in seen_slugs:
        continue
        
    seen_slugs.add(slug)
    
    try:
        structured_doc = extract_single_scheme(scheme)
        
        if structured_doc:
            structured_schemes.append(structured_doc)
            
        # Small delay to prevent IP blocking from the Govt API
        time.sleep(0.1)
        
    except Exception as e:
        print(f"Failed to process scheme {slug}: {e}")

print(f"\nExtraction Complete! Successfully processed {len(structured_schemes)} unique schemes.")
```

```python
# Save the final Master JSON
with open("govt_structured_master.json", "w", encoding="utf-8") as f:
    json.dump(
        structured_schemes, 
        f, 
        ensure_ascii=False, 
        indent=2
    )

print("Saved to govt_structured_master.json")

# Print a small sample to verify structure
if structured_schemes:
    print("\nSample Output:")
    print(json.dumps(structured_schemes[0], indent=2)[:800] + "\n...")
```

```python
import json
import re
import html

# Load the structured data
with open("govt_structured_master.json", "r", encoding="utf-8") as f:
    data = json.load(f)

def deep_unescape(text):
    """
    Recursively unescapes HTML entities until the text stops changing.
    This fixes heavily nested artifacts like &amp;amp;quot;
    """
    previous = ""
    while text != previous:
        previous = text
        text = html.unescape(text)
    return text

def fix_squished_markdown(text):
    """
    Injects spaces between squished bold markdown tags.
    Converts: **Sl. No.****Components** Into:     **Sl. No.** **Components**
    """
    return re.sub(r'\*\*\*\*', '** **', text)

# Apply the deep cleaning
for scheme in data:
    for section in scheme['content']:
        raw_text = section['text']
        
        # 1. Unescape all HTML (fixes &#39; and &amp;quot;)
        clean_text = deep_unescape(raw_text)
        
        # 2. Fix the squished markdown tables
        clean_text = fix_squished_markdown(clean_text)
        
        # 3. Clean up non-breaking spaces (\u00a0) and zero-width spaces (\u200b)
        clean_text = clean_text.replace('\u00a0', ' ').replace('\u200b', '')
        
        section['text'] = clean_text

# Save it back out
with open("govt_structured_master.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("Deep cleanup complete! All HTML entities and squished Markdown have been fixed.")
```

```python
# Audit Script to find lingering HTML-like artifacts
suspicious_pattern = re.compile(r'&\w+;|&#\d+;')
found_issues = False

for scheme in data:
    for section in scheme['content']:
        matches = suspicious_pattern.findall(section['text'])
        if matches:
            found_issues = True
            print(f"Suspicious text found in {scheme['doc_id']}: {matches}")

if not found_issues:
    print("Audit passed: No lingering HTML entities found!")
```
